```python
import torch
from torch.utils.data import Dataset, DataLoader
from transformers import ViTFeatureExtractor, GPT2Tokenizer, GPT2LMHeadModel
from PIL import Image

class CustomImageCaptioningDataset(Dataset):
    def __init__(self, annotations, feature_extractor, tokenizer, max_length=512):
        self.annotations = annotations
        self.feature_extractor = feature_extractor
        self.tokenizer = tokenizer
        # Set pad token
        self.tokenizer.pad_token = self.tokenizer.eos_token
        self.max_length = max_length

    def __len__(self):
        return len(self.annotations)

    def __getitem__(self, idx):
        item = self.annotations[idx]
        img_path = os.path.join(self.img_dir, item['image_path'])
        image = Image.open(img_path).convert('RGB')
        inputs = self.feature_extractor(images=image, return_tensors="pt")

        caption = item['description']
        # Tokenize with padding
        encoded_caption = self.tokenizer(caption, padding='max_length', max_length=self.max_length, truncation=True, return_tensors="pt")
        
        return inputs, encoded_caption.input_ids.squeeze()

    
feature_extractor = ViTFeatureExtractor.from_pretrained('google/vit-base-patch16-224')
# Load the tokenizer
tokenizer = GPT2Tokenizer.from_pretrained('gpt2')

# Set the EOS token as the padding token
tokenizer.pad_token = tokenizer.eos_token
model = GPT2LMHeadModel.from_pretrained('gpt2')  # Replace with your specific model

data = DESCRIPTION_DATA
# Create the dataset and dataloader
art_dataset = ArtDataset(data, feature_extractor, tokenizer)
art_dataloader = DataLoader(art_dataset, batch_size=8, shuffle=True)
```
