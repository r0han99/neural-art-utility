```python
def fabricate_link(url):
    name = url.split('/')[-1:][0]
    try:
        wiki_dfs = pd.read_html(url,extract_links='all')
        df = wiki_dfs[0]
        df.columns = range(len(df.columns))
        crude_link = df[1][0][1]
        image_endpoint = crude_link.split('/')
        pattern = f'''https://en.wikipedia.org/wiki/{name}#/media/{image_endpoint[-1:][0]}'''
        
        print(pattern)
        return pattern
    
    except:
        
        print(None)
        return 'Not Found'

    
```

```python
IMAGES = {}
for name, link in zip(meta['name'], meta['wikipedia']):
    img = fabricate_link(link)
    IMAGES[name] = img
```

```python
image_df = pd.DataFrame(pd.Series(IMAGES)).reset_index()
image_df.columns = ['name','img']
image_df
```

```python
artists_crude_links = image_df[image_df['img'] != 'Not Found']
artists_crude_links['img']
```

```python
import requests
from bs4 import BeautifulSoup
def create_img_links(url):
    response = requests.get(url)
    if response.status_code == 200:
        soup = BeautifulSoup(response.text, 'html.parser')
        meta_tags = soup.find_all('meta', {'property': 'og:image'})
        image_urls = [meta['content'] for meta in meta_tags]
        return image_urls[:1][0]
    else:
        return None
    
```

```python
artists_crude_links['image_links'] = artists_crude_links['img'].apply(create_img_links)
```

```python
artists_crude_links.to_csv('artist_image_links.csv',index=False)
```
