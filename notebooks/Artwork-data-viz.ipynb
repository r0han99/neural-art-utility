### Time Line Inspection

```python
meta
```

```python
timeline = meta[['name','years']]
timeline['years'] = timeline['years'].apply(lambda x: str(x).replace('–','-'))
timeline['start'] = timeline['years'].apply(lambda x: x.split(' - ')[0])
timeline['end'] = timeline['years'].apply(lambda x: x.split(' - ')[1])
timeline = timeline[['name','start','end']]
```

```python
timeline = timeline.sort_values(by='start').reset_index(drop=True)
timeline
```

```python

timeline.start = timeline['start'].astype(int)
timeline.end = timeline['end'].astype(int)
# Define your time intervals
periods = [
    (1200, 1300),
    (1300, 1400),
    (1400, 1500),
    (1500, 1600)
]

# Create a new column 'period' to store the assigned period for each artist
timeline['period'] = None

# Iterate through the rows and assign a period based on the start and end years
for i, row in timeline.iterrows():
    start_year = row['start']
    end_year = row['end']
    
    for period_start, period_end in periods:
        if start_year >= period_start and end_year <= period_end:
            timeline.at[i, 'period'] = f'{period_start}-{period_end}'
            break

# Print the updated DataFrame
print(timeline[['name', 'period']])
```

```python
timeline_data = []
for row in timeline.iterrows():
    t_dict = {}
    row_data = row[1]
    t_dict['name'] = row_data['name']
    t_dict['startTime'] = row_data.start
    t_dict['endTime'] = row_data.end

    timeline_data.append(t_dict)
timeline_data
```
